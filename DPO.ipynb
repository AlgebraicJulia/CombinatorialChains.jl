```julia
using Catlab, Catlab.Theories, Catlab.Graphs, Catlab.CategoricalAlgebra, Catlab.CategoricalAlgebra.CSets, Catlab.Graphics
```

```julia
using Catlab.CategoricalAlgebra.DPO
```

```julia
L = @acset Graph begin
  V=3
  E=2
  src=[1,1]
  tgt=[2,3]
end
to_graphviz(L, node_labels=true)
```

```julia
I = @acset Graph begin
  V=2
  E=1
  src=[1]
  tgt=[2]
end
to_graphviz(I, node_labels=true)
```

```julia
R = @acset Graph begin
  V=2
  E=2
  src=[1,2]
  tgt=[2,1]
end
to_graphviz(R, node_labels=true)
```

```julia
G = @acset Graph begin
  V=4
  E=3
  src=[2,2,2]
  tgt=[1,3,4]
end
to_graphviz(G, node_labels=true)
```

```julia
l = ACSetTransformation(I, L, V=[1,2], E=[1]);
```

```julia
r = ACSetTransformation(I, R, V=[1,2], E=[1]);
m = ACSetTransformation(L, G, V=[2, 3, 4], E=[2,3]);
```

```julia
H = rewrite_match(l, r, m)
to_graphviz(H, node_labels=true)
```

```julia
H2 = rewrite(l, r, G)
to_graphviz(H2, node_labels=true)
```
